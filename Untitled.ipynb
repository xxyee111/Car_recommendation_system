```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from typing import Dict, Tuple, Optional
import warnings
warnings.filterwarnings('ignore')
```

```python
class CarRecommendationSystem:
    
    #第一步：设置初始属性
    def __init__(self):
        self.model = None                      #训练成熟的模型
        self.label_encoder = None              #标签
        self.feature_columns = None            #特征
        self.car_mapping = None                #储存映射关系
        self.is_trained = False                #标记模型是否完成训练
        self.car_info_dict = {                 
            'byd_dolphin': {
                'name': '比亚迪海豚',
                'price_range': '10-15万',
                'description': '经济实用的纯电小车，适合城市通勤',
                'type': '紧凑型轿车'
            },
            'toyota_corolla': {
                'name': '丰田卡罗拉',
                'price_range': '12-16万',
                'description': '省心耐用的家用轿车，油耗低',
                'type': '紧凑型轿车'
            },
            'LYNK&CO_03': {
                'name': '领克03',
                'price_range': '15-20万',
                'description': '运动时尚，适合年轻人',
                'type': '紧凑型轿车'
            },
            'Honda_CR-V': {
                'name': '本田CR-V',
                'price_range': '18-27万',
                'description': '舒适的家用SUV，空间大',
                'type': '紧凑型SUV'
            },
            'Audi_A4L': {
                'name': '奥迪A4L',
                'price_range': '30-40万',
                'description': '豪华商务轿车，品质出众',
                'type': '中型轿车'
            },
            'Mercedes_E': {
                'name': '奔驰E级',
                'price_range': '45-65万',
                'description': '顶级商务座驾，彰显身份',
                'type': '中大型轿车'
            },
            'Toyota_Highlander': {
                'name': '丰田汉兰达',
                'price_range': '25-35万',
                'description': '7座大空间SUV，全家出行首选',
                'type': '中型SUV'
            },
            'tank_300': {
                'name': '坦克300',
                'price_range': '20-30万',
                'description': '硬派越野SUV，通过性强',
                'type': '中型SUV'
            },
            'WulingHongguang': {
                'name': '五菱宏光',
                'price_range': '5-10万',
                'description': '实用工具车，载人拉货两相宜',
                'type': 'MPV'
            }
        }
    
    #加载训练数据
    def load_data(self, file_path: str, encoding: str = 'gbk') -> pd.DataFrame:     #指定函数返回类型
        try:
            df = pd.read_csv(file_path, encoding=encoding)
            print(f"数据加载成功")
            return df
        except Exception as e:
            print(f"数据加载失败")
            raise                    #显示报错

    
    #数据预处理：特征工程
    def prepare_features(self, df: pd.DataFrame) -> Tuple[pd.DataFrame, np.ndarray]:    
    
        # 基础特征
        basic_features = ['budget', 'passenger', 'using']
        
        # 创建交互特征，增加对特征关系的捕捉
        df['budget_passenger'] = df['budget'] * df['passenger']
        df['budget_purpose'] = df['budget'] * df['using']
        df['passenger_purpose'] = df['passenger'] * df['using']
        self.feature_columns = basic_features + ['budget_passenger', 'budget_purpose', 'passenger_purpose']
        X = df[self.feature_columns]
        
        # 编码标签，创建1-10映射
        self.label_encoder = LabelEncoder()
        y = self.label_encoder.fit_transform(df['recommendation'])
        
        self.car_mapping = dict(zip(
            self.label_encoder.classes_,
            self.label_encoder.transform(self.label_encoder.classes_)
        ))
        
        print(f"\n✓ 特征工程完成")
        print(f"  特征数量: {len(self.feature_columns)}")
        print(f"  车型类别: {len(self.car_mapping)}")
        
        return X, y

    #训练函数
    def train(self, X: pd.DataFrame, y: np.ndarray, 
              test_size: float = 0.15, 
              random_state: int = 42) -> Dict:

        # 划分数据集
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, 
            test_size=test_size,
            random_state=random_state,
            stratify=y
        )
        
        print(f"\n 数据集划分完成")
        print(f"  训练集: {X_train.shape}")
        print(f"  测试集: {X_test.shape}")
        
        # 初始化模型
        self.model = RandomForestClassifier(
            n_estimators=150,
            max_depth=20,
            min_samples_split=5,
            min_samples_leaf=2,
            random_state=random_state,
            n_jobs=-1
        )
        
        # 训练
        self.model.fit(X_train, y_train)
        
        # 预测
        y_pred = self.model.predict(X_test)
        
        # 评估
        accuracy = accuracy_score(y_test, y_pred)
        
        # 交叉验证
        cv_scores = cross_val_score(self.model, X_train, y_train, cv=5)
        
        self.is_trained = True
        
        results = {
            'accuracy': accuracy,
            'cv_mean': cv_scores.mean(),
            'cv_std': cv_scores.std(),
            'X_test': X_test,
            'y_test': y_test,
            'y_pred': y_pred
        }
        
        print(f"\n 模型训练完成！")
        print(f"  测试集准确率: {accuracy:.2%}")
        print(f"  交叉验证均值: {cv_scores.mean():.2%}")
        print(f"  交叉验证标准差: ±{cv_scores.std()*2:.2%}")
        
        return results

    #测评模型准确性
    def evaluate(self, y_test: np.ndarray, y_pred: np.ndarray) -> None:
    
    
         print("\n模型评估报告")
         print("=" * 40)
    
         #  打印分类报告
         print("\n 分类报告:")
         print(classification_report(y_test, y_pred))
    
         #  打印特征重要性
         print("\n 特征重要性:")
         for feature, importance in zip(self.feature_columns, self.model.feature_importances_):
              print(f"  {feature}: {importance:.4f}")

    #混淆矩阵
    def plot_confusion_matrix(self, y_test: np.ndarray, y_pred: np.ndarray) -> None:
        plt.figure(figsize=(12, 10))
        cm = confusion_matrix(y_test, y_pred)
        
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=self.label_encoder.classes_,
                    yticklabels=self.label_encoder.classes_)
        
        plt.title('Confusion Matrix - Car Recommendation', fontsize=16, pad=20)
        plt.xlabel('Predicted Car', fontsize=12)
        plt.ylabel('True Car', fontsize=12)
        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight')

    #预测函数
    def predict_single(self, budget: int, passenger: int, using: int) -> Dict:
        
        if not self.is_trained:
            raise ValueError("模型未训练，请先调用 train() 方法")
        
        # 构造输入特征
        input_data = pd.DataFrame({
            'budget': [budget],
            'passenger': [passenger],
            'using': [using],
            'budget_passenger': [budget * passenger],
            'budget_purpose': [budget * using],
            'passenger_purpose': [passenger * using]
        })
        
        # 预测
        prediction_encoded = self.model.predict(input_data)[0]
        probabilities = self.model.predict_proba(input_data)[0]
        
        # 解码
        car_name = self.label_encoder.inverse_transform([prediction_encoded])[0]
        confidence = probabilities[prediction_encoded]
        
        # 获取车型信息
        car_info = self.car_info_dict.get(car_name, {})
        
        return {
            'car_name': car_name,
            'display_name': car_info.get('name', car_name),
            'confidence': confidence,
            'price_range': car_info.get('price_range', '暂无信息'),
            'description': car_info.get('description', '暂无描述'),
            'type': car_info.get('type', '未知类型')
        }

    #保存模型
    def save_model(self, file_path: str = 'car_recommendation_model.pkl') -> None:
        
        if not self.is_trained:
            raise ValueError("模型未训练，无法保存")
        
        model_data = {
            'model': self.model,
            'car_info_dict': self.car_info_dict,
            'label_encoder': self.label_encoder,
            'feature_columns': self.feature_columns,
            'car_mapping': self.car_mapping
        }
        
        joblib.dump(model_data, file_path)
        print(f"\n 模型已保存至: {file_path}")

    #加载模型
    def load_model(self, file_path: str = 'car_recommendation_model.pkl') -> None:

        model_data = joblib.load(file_path)
        
        self.model = model_data['model']
        self.label_encoder = model_data['label_encoder']
        self.feature_columns = model_data['feature_columns']
        self.car_mapping = model_data['car_mapping']
        self.car_info_dict = model_data['car_info_dict']
        self.is_trained = True
   
```

```python
# 测试代码
def main():
    try:
        print("="*50)
        print("汽车推荐系统训练流程")
        print("="*50)
        
        # 创建实例
        system = CarRecommendationSystem()
        
        # 1. 加载数据
        df = system.load_data('training_data.csv', encoding='gbk')
        
        # 2. 准备特征
        X, y = system.prepare_features(df)
        
        # 3. 训练模型
        results = system.train(X, y)
        
        # 4. 评估模型
        system.evaluate(results['y_test'], results['y_pred'])
        
        # 5. 可视化
        system.plot_confusion_matrix(results['y_test'], results['y_pred'])
        
        # 6. 保存模型
        system.save_model()
        
        print("\n" + "="*50)
        print("✓ 训练完成！现在可以使用模型进行推荐了")
        print("="*50)
        
        # 测试预测
        print("\n测试预测功能:")
        test_result = system.predict_single(budget=2, passenger=2, using=2)
        print(f"  推荐车型: {test_result['display_name']}")
        print(f"  价格区间: {test_result['price_range']}")
        print(f"  置信度: {test_result['confidence']:.2%}")
        
    except Exception as e:
        print(f"\n✗ 训练过程出错: {e}")
        import traceback
        traceback.print_exc()


if __name__ == "__main__":
    main()
```

```python
<!DOCTYPE html>
<html lang="zh-CN">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Heteroscedasticity 深度解析</title>
    <style>
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }

        body {
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
            overflow: hidden;
        }

        .presentation {
            width: 100vw;
            height: 100vh;
            display: flex;
            align-items: center;
            justify-content: center;
        }

        .slide {
            width: 90%;
            max-width: 1200px;
            height: 85vh;
            background: white;
            border-radius: 20px;
            box-shadow: 0 20px 60px rgba(0,0,0,0.3);
            padding: 60px;
            display: none;
            flex-direction: column;
            position: relative;
            animation: slideIn 0.5s ease-out;
        }

        .slide.active {
            display: flex;
        }

        @keyframes slideIn {
            from {
                opacity: 0;
                transform: translateY(30px);
            }
            to {
                opacity: 1;
                transform: translateY(0);
            }
        }

        h1 {
            font-size: 3.5em;
            color: #667eea;
            margin-bottom: 20px;
            text-align: center;
        }

        h2 {
            font-size: 2.5em;
            color: #667eea;
            margin-bottom: 30px;
            border-bottom: 4px solid #667eea;
            padding-bottom: 15px;
        }

        h3 {
            font-size: 1.8em;
            color: #764ba2;
            margin: 25px 0 15px 0;
        }

        .subtitle {
            font-size: 1.5em;
            color: #666;
            text-align: center;
            margin-bottom: 40px;
        }

        .content {
            flex: 1;
            overflow-y: auto;
            font-size: 1.2em;
            line-height: 1.8;
            color: #333;
        }

        .pronunciation {
            background: #f0f4ff;
            padding: 20px;
            border-radius: 10px;
            font-size: 1.3em;
            text-align: center;
            margin: 20px 0;
            border-left: 5px solid #667eea;
        }

        .definition-box {
            background: linear-gradient(135deg, #667eea15 0%, #764ba215 100%);
            padding: 25px;
            border-radius: 15px;
            margin: 20px 0;
            border-left: 5px solid #764ba2;
        }

        .comparison {
            display: grid;
            grid-template-columns: 1fr 1fr;
            gap: 20px;
            margin: 20px 0;
        }

        .comparison-item {
            background: #f8f9fa;
            padding: 20px;
            border-radius: 10px;
            border: 2px solid #ddd;
        }

        .comparison-item h4 {
            color: #667eea;
            margin-bottom: 10px;
            font-size: 1.3em;
        }

        .chart-container {
            display: flex;
            justify-content: space-around;
            margin: 30px 0;
            gap: 20px;
        }

        .chart {
            flex: 1;
            text-align: center;
        }

        canvas {
            border: 2px solid #ddd;
            border-radius: 10px;
            background: white;
        }

        .formula {
            background: #f8f9fa;
            padding: 20px;
            border-radius: 10px;
            font-family: 'Courier New', monospace;
            font-size: 1.1em;
            margin: 15px 0;
            text-align: center;
            border: 2px dashed #667eea;
        }

        ul {
            margin-left: 30px;
            margin-top: 15px;
        }

        li {
            margin: 12px 0;
            padding-left: 10px;
        }

        .highlight {
            background: #fff59d;
            padding: 2px 6px;
            border-radius: 3px;
            font-weight: bold;
        }

        .controls {
            position: fixed;
            bottom: 30px;
            left: 50%;
            transform: translateX(-50%);
            display: flex;
            gap: 15px;
            z-index: 1000;
        }

        button {
            padding: 15px 30px;
            font-size: 1.1em;
            border: none;
            border-radius: 30px;
            background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
            color: white;
            cursor: pointer;
            transition: all 0.3s;
            box-shadow: 0 5px 15px rgba(0,0,0,0.2);
        }

        button:hover {
            transform: translateY(-3px);
            box-shadow: 0 8px 20px rgba(0,0,0,0.3);
        }

        button:disabled {
            opacity: 0.5;
            cursor: not-allowed;
            transform: none;
        }

        .slide-number {
            position: absolute;
            bottom: 20px;
            right: 30px;
            color: #999;
            font-size: 1em;
        }

        .example-box {
            background: #e8f5e9;
            padding: 20px;
            border-radius: 10px;
            margin: 15px 0;
            border-left: 5px solid #4caf50;
        }
    </style>
</head>
<body>
    <div class="presentation">
        <!-- Slide 1: Title -->
        <div class="slide active">
            <h1>Heteroscedasticity</h1>
            <div class="subtitle">异方差性：回归分析中的关键概念</div>
            <div class="content">
                <div style="text-align: center; margin-top: 80px;">
                    <svg width="400" height="300" style="margin: 0 auto;">
                        <line x1="50" y1="250" x2="350" y2="250" stroke="#667eea" stroke-width="2"/>
                        <line x1="50" y1="250" x2="50" y2="50" stroke="#667eea" stroke-width="2"/>
                        
                        <!-- Scatter points with increasing variance -->
                        <circle cx="80" cy="230" r="3" fill="#764ba2" opacity="0.6"/>
                        <circle cx="85" cy="235" r="3" fill="#764ba2" opacity="0.6"/>
                        
                        <circle cx="140" cy="190" r="3" fill="#764ba2" opacity="0.6"/>
                        <circle cx="145" cy="210" r="3" fill="#764ba2" opacity="0.6"/>
                        <circle cx="150" cy="185" r="3" fill="#764ba2" opacity="0.6"/>
                        
                        <circle cx="200" cy="140" r="3" fill="#764ba2" opacity="0.6"/>
                        <circle cx="210" cy="170" r="3" fill="#764ba2" opacity="0.6"/>
                        <circle cx="205" cy="120" r="3" fill="#764ba2" opacity="0.6"/>
                        <circle cx="215" cy="160" r="3" fill="#764ba2" opacity="0.6"/>
                        
                        <circle cx="270" cy="80" r="3" fill="#764ba2" opacity="0.6"/>
                        <circle cx="280" cy="130" r="3" fill="#764ba2" opacity="0.6"/>
                        <circle cx="275" cy="60" r="3" fill="#764ba2" opacity="0.6"/>
                        <circle cx="285" cy="110" r="3" fill="#764ba2" opacity="0.6"/>
                        <circle cx="290" cy="90" r="3" fill="#764ba2" opacity="0.6"/>
                        <circle cx="265" cy="100" r="3" fill="#764ba2" opacity="0.6"/>
                        
                        <!-- Trend line -->
                        <line x1="60" y1="240" x2="330" y2="70" stroke="#ff6b6b" stroke-width="2" stroke-dasharray="5,5"/>
                        
                        <text x="200" y="280" text-anchor="middle" fill="#667eea" font-size="14">X (自变量)</text>
                        <text x="30" y="150" text-anchor="middle" fill="#667eea" font-size="14" transform="rotate(-90 30 150)">Y (因变量)</text>
                    </svg>
                    <p style="margin-top: 30px; font-size: 1.1em; color: #666;">
                        误差的方差随自变量变化而变化
                    </p>
                </div>
            </div>
            <div class="slide-number">1 / 8</div>
        </div>

        <!-- Slide 2: Pronunciation & Etymology -->
        <div class="slide">
            <h2>📢 发音与词源</h2>
            <div class="content">
                <div class="pronunciation">
                    <strong>发音：</strong> /ˌhetərəʊskəˈdæstɪsəti/<br>
                    <strong>中文读音：</strong> 黑特罗斯克达斯提西提
                </div>

                <h3>词源分析</h3>
                <div class="definition-box">
                    <p><strong>Hetero-</strong> (希腊语) = 不同的、异质的</p>
                    <p><strong>-scedastic</strong> (希腊语 skedasis) = 分散、散布</p>
                    <p><strong>-ity</strong> = 名词后缀，表示状态或性质</p>
                </div>

                <div style="margin-top: 30px; padding: 20px; background: #fff3e0; border-radius: 10px;">
                    <h3 style="color: #f57c00;">💡 字面意思</h3>
                    <p style="font-size: 1.3em; margin-top: 15px;">
                        <strong>"不同的分散程度"</strong> — 指的是数据的离散程度（方差）不是恒定的，而是变化的。
                    </p>
                </div>

                <h3 style="margin-top: 30px;">对应概念</h3>
                <p style="font-size: 1.1em; margin-top: 10px;">
                    <strong>Homoscedasticity</strong> (同方差性): <em>Homo-</em> = 相同的<br>
                    表示误差的方差在所有观测点上保持恒定
                </p>
            </div>
            <div class="slide-number">2 / 8</div>
        </div>

        <!-- Slide 3: Definition -->
        <div class="slide">
            <h2>📖 定义与含义</h2>
            <div class="content">
                <div class="definition-box">
                    <h3 style="margin-top: 0;">统计学定义</h3>
                    <p style="font-size: 1.2em; line-height: 1.8;">
                        <strong>异方差性</strong>是指在回归分析中，<span class="highlight">误差项的方差不是常数</span>，
                        而是随着自变量的变化而系统性地变化。
                    </p>
                </div>

                <h3>数学表达</h3>
                <div class="formula">
                    Var(εᵢ) = σᵢ² ≠ σ² (常数)
                </div>
                <p style="margin-top: 10px; color: #666;">
                    其中 εᵢ 是第 i 个观测的误差项，σᵢ² 是其方差
                </p>

                <h3>核心特征</h3>
                <ul style="font-size: 1.1em;">
                    <li><strong>方差非恒定：</strong>不同观测点的误差方差不同</li>
                    <li><strong>系统性变化：</strong>方差通常与一个或多个自变量相关</li>
                    <li><strong>违反假设：</strong>违反了经典线性回归的基本假设之一</li>
                    <li><strong>影响推断：</strong>会导致标准误估计偏误，影响假设检验</li>
                </ul>

                <div class="example-box">
                    <h4 style="color: #2e7d32; margin-bottom: 10px;">🔍 直观理解</h4>
                    <p>想象射击靶子：如果距离越远，弹孔的分散程度越大，这就是异方差性。
                    在回归中，随着 X 增大，Y 的预测误差可能越来越分散。</p>
                </div>
            </div>
            <div class="slide-number">3 / 8</div>
        </div>

        <!-- Slide 4: Visual Comparison -->
        <div class="slide">
            <h2>📊 视觉对比</h2>
            <div class="content">
                <div class="comparison">
                    <div class="comparison-item">
                        <h4>✅ 同方差性 (Homoscedasticity)</h4>
                        <canvas id="homoCanvas" width="250" height="200"></canvas>
                        <p style="margin-top: 10px;">误差的分散程度恒定</p>
                    </div>
                    <div class="comparison-item">
                        <h4>⚠️ 异方差性 (Heteroscedasticity)</h4>
                        <canvas id="heteroCanvas" width="250" height="200"></canvas>
                        <p style="margin-top: 10px;">误差的分散程度递增</p>
                    </div>
                </div>

                <h3 style="margin-top: 30px;">残差图特征</h3>
                <div class="definition-box">
                    <p><strong>同方差：</strong>残差图呈现随机散布，无明显模式</p>
                    <p><strong>异方差：</strong>残差图呈现"喇叭形"或其他系统性模式</p>
                </div>

                <div style="margin-top: 20px; padding: 15px; background: #ffebee; border-radius: 10px; border-left: 5px solid #f44336;">
                    <strong>⚠️ 检验方法：</strong>
                    <ul style="margin-top: 10px;">
                        <li>视觉检查：残差图、残差平方图</li>
                        <li>Breusch-Pagan 检验</li>
                        <li>White 检验</li>
                        <li>Goldfeld-Quandt 检验</li>
                    </ul>
                </div>
            </div>
            <div class="slide-number">4 / 8</div>
        </div>

        <!-- Slide 5: Historical Origin -->
        <div class="slide">
            <h2>🏛️ 历史来源</h2>
            <div class="content">
                <h3>发展时间线</h3>
                <div style="margin: 20px 0;">
                    <div style="padding: 15px; margin: 15px 0; background: #e3f2fd; border-left: 5px solid #2196f3; border-radius: 5px;">
                        <strong>19世纪末：</strong>高斯-马尔科夫定理确立，假设同方差性
                    </div>
                    <div style="padding: 15px; margin: 15px 0; background: #f3e5f5; border-left: 5px solid #9c27b0; border-radius: 5px;">
                        <strong>1950年代：</strong>计量经济学兴起，异方差问题受到关注
                    </div>
                    <div style="padding: 15px; margin: 15px 0; background: #fff3e0; border-left: 5px solid #ff9800; border-radius: 5px;">
                        <strong>1970-1980年代：</strong>White (1980) 提出稳健标准误，Breusch-Pagan 开发检验方法
                    </div>
                    <div style="padding: 15px; margin: 15px 0; background: #e8f5e9; border-left: 5px solid #4caf50; border-radius: 5px;">
                        <strong>现代：</strong>成为回归分析的标准诊断内容
                    </div>
                </div>

                <h3>关键贡献者</h3>
                <ul style="font-size: 1.1em;">
                    <li><strong>Halbert White (1980):</strong> 异方差稳健标准误（White标准误）</li>
                    <li><strong>Trevor Breusch & Adrian Pagan (1979):</strong> BP检验</li>
                    <li><strong>Stephen Goldfeld & Richard Quandt (1965):</strong> GQ检验</li>
                </ul>

                <div class="example-box" style="margin-top: 20px;">
                    <h4 style="color: #2e7d32;">📚 学术影响</h4>
                    <p>异方差性的研究推动了稳健统计方法的发展，是现代计量经济学的基石之一。</p>
                </div>
            </div>
            <div class="slide-number">5 / 8</div>
        </div>

        <!-- Slide 6: Common Causes -->
        <div class="slide">
            <h2>🔍 产生原因</h2>
            <div class="content">
                <h3>1. 数据特性</h3>
                <ul>
                    <li><strong>规模差异：</strong>大企业和小企业的财务数据方差差异</li>
                    <li><strong>学习效应：</strong>随着经验增加，表现的稳定性提高</li>
                    <li><strong>测量误差：</strong>不同范围的测量精度不同</li>
                </ul>

                <div class="example-box">
                    <h4 style="color: #2e7d32;">💡 经典例子</h4>
                    <p><strong>收入与消费：</strong>高收入家庭的消费模式更多样化，方差更大；
                    低收入家庭消费更集中在必需品，方差较小。</p>
                </div>

                <h3>2. 模型设定</h3>
                <ul>
                    <li><strong>遗漏变量：</strong>未包含的重要变量导致误差项包含更多信息</li>
                    <li><strong>错误的函数形式：</strong>线性模型拟合非线性关系</li>
                    <li><strong>异常值影响：</strong>极端值导致局部方差增大</li>
                </ul>

                <h3>3. 数据收集过程</h3>
                <ul>
                    <li><strong>抽样方法：</strong>分层抽样可能导致不同层的方差不同</li>
                    <li><strong>时间聚合：</strong>不同时间段的数据波动性不同</li>
                    <li><strong>空间异质性：</strong>不同地区的特征差异</li>
                </ul>
            </div>
            <div class="slide-number">6 / 8</div>
        </div>

        <!-- Slide 7: Consequences & Solutions -->
        <div class="slide">
            <h2>⚡ 影响与处理</h2>
            <div class="content">
                <h3>🔴 主要影响</h3>
                <div class="comparison">
                    <div class="comparison-item" style="border-color: #f44336;">
                        <h4>估计方面</h4>
                        <ul style="margin-left: 20px; font-size: 0.95em;">
                            <li>OLS估计仍然无偏</li>
                            <li>但不再是最优线性无偏估计</li>
                            <li>失去效率</li>
                        </ul>
                    </div>
                    <div class="comparison-item" style="border-color: #ff9800;">
                        <h4>推断方面</h4>
                        <ul style="margin-left: 20px; font-size: 0.95em;">
                            <li>标准误估计有偏</li>
                            <li>t统计量和F统计量失效</li>
                            <li>置信区间不准确</li>
                        </ul>
                    </div>
                </div>

                <h3 style="margin-top: 25px;">✅ 解决方法</h3>
                
                <div style="background: #e8f5e9; padding: 20px; border-radius: 10px; margin: 15px 0;">
                    <h4 style="color: #2e7d32; margin-bottom: 10px;">1️⃣ 稳健标准误 (Robust Standard Errors)</h4>
                    <p>使用 White 异方差稳健标准误，不改变系数估计，只修正标准误</p>
                </div>

                <div style="background: #e3f2fd; padding: 20px; border-radius: 10px; margin: 15px 0;">
                    <h4 style="color: #1976d2; margin-bottom: 10px;">2️⃣ 加权最小二乘法 (WLS)</h4>
                    <p>如果知道方差结构，给予低方差观测更高权重</p>
                    <div class="formula" style="font-size: 0.9em; margin-top: 10px;">
                        权重 wᵢ = 1/σᵢ²
                    </div>
                </div>

                <div style="background: #fff3e0; padding: 20px; border-radius: 10px; margin: 15px 0;">
                    <h4 style="color: #f57c00; margin-bottom: 10px;">3️⃣ 变量转换</h4>
                    <p>对数转换、平方根转换等，可能稳定方差</p>
                </div>

                <div style="background: #f3e5f5; padding: 20px; border-radius: 10px; margin: 15px 0;">
                    <h4 style="color: #7b1fa2; margin-bottom: 10px;">4️⃣ 改进模型设定</h4>
                    <p>添加遗漏变量、改变函数形式、处理异常值</p>
                </div>
            </div>
            <div class="slide-number">7 / 8</div>
        </div>

        <!-- Slide 8: Applications -->
        <div class="slide">
            <h2>🌍 实际应用</h2>
            <div class="content">
                <h3>经济学领域</h3>
                <div class="example-box">
                    <p><strong>收入研究：</strong>高收入群体消费行为的方差通常大于低收入群体</p>
                    <p><strong>股票收益：</strong>不同市值公司的收益波动性差异显著</p>
                    <p><strong>房地产价格：</strong>豪宅价格波动大于普通住宅</p>
                </div>

                <h3>金融领域</h3>
                <div class="example-box">
                    <p><strong>风险建模：</strong>ARCH/GARCH模型处理时变波动率</p>
                    <p><strong>投资组合：</strong>不同资产的收益方差随市场条件变化</p>
                </div>

                <h3>其他领域</h3>
                <ul style="font-size: 1.1em;">
                    <li><strong>医学研究：</strong>治疗效果在不同病情严重程度患者中的差异</li>
                    <li><strong>教育研究：</strong>学生成绩方差随教学投入变化</li>
                    <li><strong>环境科学：</strong>污染物浓度的测量误差随浓度水平变化</li>
                    <li><strong>工程学：</strong>产品质量随生产批次大小变化</li>
                </ul>

                <div style="margin-top: 30px; padding: 25px; background: linear-gradient(135deg, #667eea15 0%, #764ba215 100%); border-radius: 15px; border: 3px solid #667eea;">
                    <h3 style="color: #667eea; text-align: center; margin-bottom: 15px;">💡 关键要点</h3>
                    <p style="font-size: 1.2em; text-align: center; line-height: 1.8;">
                        <strong>异方差性不是"错误"，而是数据的真实特征。</strong><br>
                        正确识别和处理异方差性是进行可靠统计推断的关键！
                    </p>
                </div>
            </div>
            <div class="slide-number">8 / 8</div>
        </div>
    </div>

    <div class="controls">
        <button id="prevBtn" onclick="changeSlide(-1)">← 上一页</button>
        <button id="nextBtn" onclick="changeSlide(1)">下一页 →</button>
    </div>

    <script>
        let currentSlide = 0;
        const slides = document.querySelectorAll('.slide');
        const prevBtn = document.getElementById('prevBtn');
        const nextBtn = document.getElementById('nextBtn');

        function showSlide(n) {
            slides[currentSlide].classList.remove('active');
            currentSlide = (n + slides.length) % slides.length;
            slides[currentSlide].classList.add('active');
            
            prevBtn.disabled = currentSlide === 0;
            nextBtn.disabled = currentSlide === slides.length - 1;

            // Draw charts when showing slide 4
            if (currentSlide === 3) {
                drawCharts();
            }
        }

        function changeSlide(direction) {
            showSlide(currentSlide + direction);
        }

        // Keyboard navigation
        document.addEventListener('keydown', (e) => {
            if (e.key === 'ArrowLeft') changeSlide(-1);
            if (e.key === 'ArrowRight') changeSlide(1);
        });

        // Draw comparison charts
        function drawCharts() {
            // Homoscedastic chart
            const homoCanvas = document.getElementById('homoCanvas');
            const homoCtx = homoCanvas.getContext('2d');
            
            homoCtx.clearRect(0, 0, homoCanvas.width, homoCanvas.height);
            
            // Axes
            homoCtx.strokeStyle = '#667eea';
            homoCtx.lineWidth = 2;
            homoCtx.beginPath();
            homoCtx.moveTo(30, 170);
            homoCtx.lineTo(230, 170);
            homoCtx.moveTo(30, 170);
            homoCtx.lineTo(30, 30);
            homoCtx.stroke();
            
            // Regression line
            homoCtx.strokeStyle = '#ff6b6b';
            homoCtx.setLineDash([5, 5]);
            homoCtx.beginPath();
            homoCtx.moveTo(40, 160);
            homoCtx.lineTo(220, 40);
            homoCtx.stroke();
            homoCtx.setLineDash([]);
            
            // Points with constant variance
            homoCtx.fillStyle = '#764ba2';
            for (let i = 0; i < 30; i++) {
                const x = 40 + Math.random() * 180;
                const yBase = 160 - (x - 40) * 0.67;
                const y = yBase + (Math.random() - 0.5) * 30; // Constant variance
                homoCtx.beginPath();
                homoCtx.arc(x, y, 3, 0, 2 * Math.PI);
                homoCtx.fill();
            }
            
            // Heteroscedastic chart
            const heteroCanvas = document.getElementById('heteroCanvas');
            const heteroCtx = heteroCanvas.getContext('2d');
            
            heteroCtx.clearRect(0, 0, hetero
```
